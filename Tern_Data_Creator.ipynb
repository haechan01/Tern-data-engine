```python
import requests
import PIL.Image as Image
import io
from datetime import datetime
import json

import googlemaps # pip install googlemaps
import pandas as pd # pip install pandas
import time
import openpyxl
from datetime import date
import requests
import PIL.Image as Image
import io
from datetime import datetime
import json
from nltk.tokenize import sent_tokenize, word_tokenize
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import nltk
import re
import requests
from bs4 import BeautifulSoup
```

```python
# with descriptions

def inWikipedia(url):
  request_results = requests.get(url)
  wiki_page = BeautifulSoup(request_results.text, "html.parser")
  tag = ''
  if len(wiki_page.find_all('b'))>1:
    tag = wiki_page.find_all('b')[1].text
  if tag == 'Wikipedia does not have an article with this exact name.':
    return False
  return True


def create_table(city,n):
    geocode = map_client.geocode(address=city)
    (lat, lng) = map(geocode[0]['geometry']['location'].get, ('lat', 'lng'))
    search_string = 'sights '+city
    distance = miles_to_meters(20)
    business_list = []

    response = map_client.places_nearby(
        location=(lat, lng),
        keyword=search_string,
        radius=distance
    )   

    business_list.extend(response.get('results'))
    next_page_token = response.get('next_page_token')

    while next_page_token:
        time.sleep(2)
        response = map_client.places_nearby(
            location=(lat, lng),
            keyword=search_string,
            radius=distance,
            page_token=next_page_token
        )   
        business_list.extend(response.get('results'))
        next_page_token = response.get('next_page_token')
    idx=[]
    for i in range(len(business_list)):
        num=i+n
        idx.append(num)
    df = pd.DataFrame(business_list, index= idx)
    df['googlemaps_url'] = 'https://www.google.com/maps/place/?q=place_id:' + df['place_id']
    
    del df['icon']
    del df['icon_background_color']
    del df['icon_mask_base_uri']
    del df['opening_hours']
    del df['plus_code']
    del df['reference']
    del df['scope']

    today = date.today()
    now = datetime.now()
    dt_string = now.strftime("%Y/%m/%d %H:%M:%S")
    tern_uid = 'm06mrfQB2ERTRY4NiIQ48DlO4U43'

    currentDate = today.strftime("%Y/%m/%d")
    card_uid=[]
    card_uid.extend([i + df.index[-1]+1 -60 for i in range(len(df))])
    created_at = []
    created_at.extend([dt_string for i in range(len(df))])
    dates=[]
    dates.extend([currentDate for i in range(len(df))])
    timestamps = []
    timestamps.extend([now for i in range(len(df))])
    authors=[]
    authors.extend([tern_uid]*len(df))
    likes=[]
    likes.extend([0 for i in range(len(df))])
    dislikes=[]
    dislikes.extend([0 for i in range(len(df))])
    mustDos=[]
    mustDos.extend([0 for i in range(len(df))])
    links = []
    links.extend(['N/A' for i in range(len(df))])
    geometries=[]
    geometries.extend([[df['geometry'][i+n]['location']['lat'],df['geometry'][i+n]['location']['lng']] for i in range(len(df))])
    popularities = []
    popularities.extend([(df['rating'][i+n])/5*0.75 for i in range(len(df))])
    df['last_updated'] = dates
    df['created_by'] = authors
    df['created_at'] = created_at
    df['likes'] = likes
    df['dislikes'] = dislikes
    df['mustDos'] = mustDos
    df['card_uid'] = card_uid
    df['geometries'] = geometries
    df['links'] = links
    df['timestamp'] = timestamps
    df['popularity'] = popularities
    df['top_popularity'] = popularities

    df.loc[df['business_status'] == 'OPERATIONAL', 'business_status'] = 0
    df.loc[df['business_status'] == 'CLOSED_TEMPORARILY', 'business_status'] = 1
    df.loc[df['business_status'] == 'CLOSED_PERMANENTLY', 'business_status'] = 2
    del df['geometry']
    df.rename(columns = {'vicinity':'address'}, inplace = True)
    df.rename(columns = {'types':'tags'}, inplace = True)

    clist = ['card_uid','place_id','geometries','business_status','name','photos','address','googlemaps_url','tags','last_updated','created_by','created_at','rating','user_ratings_total','likes','dislikes','mustDos', 'popularity','top_popularity','links']
    df = df[clist]


    websites=[]
    opening_hours = []
    phone_numbers = []
    photos = []
    reviews =[]

    for i in range(len(df)):
        placeId = df['place_id'][i+n]
        url = "https://maps.googleapis.com/maps/api/place/details/json?place_id="+placeId+"&fields=name%2Cformatted_phone_number%2Cwebsite%2Copening_hours%2Cphotos%2Creviews&key="+API_KEY
        payload={}
        headers = {}
        response = requests.request("GET", url, headers=headers, data=payload)
        result = response.json().get('result')
        
        
        
        if result.get('website'):
            website = result.get('website')
        else:
            website = 'No website'

        if result.get('opening_hours'):
            opening_hour = result.get('opening_hours')['weekday_text']
        else:
            opening_hour = 'Opening hours unkown'
        
        if result.get('formatted_phone_number'):
            phone_number = result.get('formatted_phone_number')
        else:
            phone_number = 'phone number unkown'
        
        if result.get('photos'):
            photo = result.get('photos')[0]['photo_reference']
        else:
            photo = 'No photo'
        
        if result.get('reviews'):
            review=[]
            
            for i in range(len(result.get('reviews'))):
                json_string = json.dumps({'author_name': result.get('reviews')[i]['author_name'], 'rating': result.get('reviews')[i]['rating'],'text':result.get('reviews')[i]['text']})
                review.append(json_string)
        else:
            opening_hour = 'No review'
        
       
        websites.append(website)
        opening_hours.append(opening_hour)
        phone_numbers.append(phone_number)
        photos.append(photo)
        reviews.append(review)
        

    
    
    df['website'] = websites
    df['opening_hours'] = opening_hours
    df['phone_number'] = phone_numbers
    df['photo_reference'] = photos
    df['reviews'] = reviews
    df['top_card'] = df['card_uid']


    photos=[]
    for i in range(len(df)):
        photo_reference = df['photo_reference'][i+n]
        if photo_reference != 'No photo':
            baseUrl= "https://maps.googleapis.com/maps/api/place/photo?maxwidth=400&photo_reference="+photo_reference+"&key="+API_KEY
            res = requests.get(baseUrl)
            photo_byte = res.content
            photos.append(bytearray(photo_byte))
        else:
            photos.append(0)
    df['photo'] = photos
    del df['photos']
  
    links = []
    for i in range(len(df['name'])):
        tokenized_name = word_tokenize(df['name'][i+n])
        url_name = '/'
        for word in tokenized_name:
            url_name+= word+'_'
        new_url_name = url_name[:-1]
        url = "https://en.wikipedia.org/wiki" + new_url_name
        if inWikipedia(url) == True:
            links.append(url)
        else:
            alt_name='='
            for word in tokenized_name:
                alt_name += word+'+'
            new_alt_name = alt_name[:-1]
            alt_url = "https://www.google.com/search?q"+new_alt_name+"&as_epq=&as_oq=&as_eq=&as_nlo=&as_nhi=&lr=lang_en&cr=&as_qdr=all&as_sitesearch=&as_occt=any&safe=images&as_filetype=&tbs="
            links.append(alt_url)
    descriptions = []
    for link in links:
        request_results = requests.get(link)
        web_page = BeautifulSoup(request_results.text, "html.parser")
        if link[11:20]=='wikipedia':
            tag = web_page.find_all('p')[1:3]
            if len(tag)>1:
                text = tag[0].text +tag[1].text
            else:
                text = 'Descripition under construction...'
        else:
            try: 
                text = web_page.find('div', class_= 'BNeawe s3v9rd AP7Wnd').text + link
            except:
                text= 'Descripition under construction...'
        descriptions.append(text)
    df['description'] = descriptions
  
    return df
```

```python
cities=['San Francisco', "Los Angeles"]
frames = []
num=0
for i in range(len(cities)):
  new_df = create_table(cities[i],num)
  num=new_df.index[-1]+1
  frames.append(new_df)
result = pd.concat(frames)
```
